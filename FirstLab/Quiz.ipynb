```python
# импортируйте библиотеки 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Загрузка данных 
df = pd.read_csv('data.csv')

#Используйте основные операции Pandas для исследования датасета
df.head()
```

```python
#Проверьте пропущенные значения: 
print("Пропущенные значения:\n", df.isnull().sum())
```

```python
# Исправление данных в столбцах "Height" и "Weight"

#Ковертировать Height из дюйма в метры : inches_to_meters = 0.0254
inches_to_meters = 0.0254
# Конвертировать Weight из фунтов в кг : lbs_to_kg = 0.453592 
lbs_to_kg = 0.453592

# Замените старые данные исправленными данными 
df['Height'] = df['Height'] * inches_to_meters
df['Weight'] = df['Weight'] * lbs_to_kg
```

```python
# Вычислите средний рост и вес.
average_height = df['Height'].mean()
average_weight = df['Weight'].mean()
print("Height :" , average_height)
print("Weight :" , average_weight)
```

```python
# Найдите максимальный и минимальный рост. 
max_height = df['Height'].max()
min_height = df['Height'].min()
print("Максимальный рост :" , max_height)
print("Минимальный рост  :" , min_height)
```

```python
# Гистограммы распределения роста и веса
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.hist(df['Height'], bins=20, color='skyblue', edgecolor='black')
plt.title('Распределение роста')

plt.subplot(1, 2, 2)
plt.hist(df['Weight'], bins=20, color='lightcoral', edgecolor='black')
plt.title('Распределение веса')

plt.show()
```

```python
# Точечный график
# Точечный график
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Height', y='Weight', data=df, hue='Gender', palette='Set1')
plt.title('Точечный график роста и веса с разделением по полу')
plt.show()
```

```python
# Ящичная диаграмма для роста с разделением по полу
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='Height', data=df, palette='pastel')
plt.title('Ящичная диаграмма роста с разделением по полу')
plt.show()
```
